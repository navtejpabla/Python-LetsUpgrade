```python
%%writefile prime_number.py
"""
Define a Range
"""
START = 1
END = 100
"""
Write the prime number code
"""
for x in range(START, END + 1):
    for y in range(2, x):
        if x % y == 0:
            break
        print(x)
```

```python
lst = list(range(1000))


def armstrngNumGen(lst):
    for x in lst:
        order = len(str(x))
        sum = 0
        temp = x
    while temp > 0:
        digit = temp % 10
        sum += digit ** order
        temp //= 10

        if x == sum:
            yield x
            break
```
